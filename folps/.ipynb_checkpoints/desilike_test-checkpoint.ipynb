```python
import matplotlib.pyplot as plt
import numpy as np

# Example: replace with your actual data
ks = theory2.k
# a, d = FOLPS_LCDM, Rept_LCDM
# g, j = FOLPS_w0wa, Rept_w0wa
a,d,g,j=b,e,h,k
# Compute percentage differences
err_LCDM = 100 * (d - a) / a
err_w0wa = 100 * (j - g) / g

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 7), sharex=True, gridspec_kw={'height_ratios':[3,1]})

# --- Top panel: power spectra ---
ax1.plot(ks, a*ks, c='b', label="FOLPS ΛCDM")
ax1.plot(ks, d*ks, c='b', ls=':', lw=2.5, label="Rept ΛCDM")
ax1.plot(ks, g*ks, c='r', label=r"FOLPS $w_0w_a$ ($w_0=-0.7,\,w_a=-0.44$)")
ax1.plot(ks, j*ks, c='r', ls=':', lw=2.5, label=r"Rept $w_0w_a$")
ax1.set_title("l=2",fontsize=24)
ax1.set_ylabel(r'$k\,P_{\ell}(k)\,[h^{-1}\,\mathrm{Mpc}]^2$', fontsize=13)
ax1.legend(fontsize=11)
ax1.grid(alpha=0.3)
# ax1.set_xscale('log')  # optional, comment out if you prefer linear scale

# --- Bottom panel: relative error (percentage) ---
ax2.axhline(0, color='k', lw=1)
ax2.plot(ks, err_LCDM, c='b', label='ΛCDM')
ax2.plot(ks, err_w0wa, c='r', label=r'$w_0w_a$')
ax2.set_xlabel(r'$k\,[h\,\mathrm{Mpc}^{-1}]$', fontsize=13)
ax2.set_ylabel(r'$\frac{\Delta P}{P_{\mathrm{FOLPS}}}\,[\%]$', fontsize=12)
ax2.legend(fontsize=10)
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
